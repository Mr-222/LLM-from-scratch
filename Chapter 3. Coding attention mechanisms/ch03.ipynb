```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
query = inputs[1]
print(query.shape)
print(inputs.shape)
attn_score2 = torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attn_score2[i] = torch.dot(x_i, query)
print(attn_score2)
```

```python
attn_weights_2_tmp = attn_score2 / attn_score2.sum()
print("Attention weights:", attn_weights_2_tmp)
print("Sum:", attn_weights_2_tmp.sum())
```

```python
def softmax_naive(x):
    return torch.exp(x) / torch.exp(x).sum(dim=0)
```

```python
attn_weights_2_naive = softmax_naive(attn_score2)
print("Attention weights:", attn_weights_2_naive)
print("Sum:", attn_weights_2_naive.sum())
```

```python
attn_weights_2 = torch.softmax(attn_score2, dim=0)
print("Attention weights:", attn_weights_2)
print("Sum:", attn_weights_2.sum())
```

```python
query = inputs[1]
context_vec_2 = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_vec_2 += attn_weights_2[i] * x_i
print(context_vec_2)
```

```python
attn_scores = torch.empty(6, 6)
for i, x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attn_scores[i, j] = torch.dot(x_i, x_j)
print(attn_scores)
```

```python
attn_scores = inputs @ inputs.T
print(attn_scores)
```

```python
attn_weights = torch.softmax(attn_scores, dim=-1)
print(attn_weights)
```

```python
print("All row sums:", attn_weights.sum(dim=-1))
```

```python
all_context_vecs = attn_weights @ inputs
print(all_context_vecs)
```

```python
d_k = keys.shape[-1]
attn_weights_2 = torch.softmax(attn_score2 / d_k ** 0.5, dim=-1)
print(attn_weights_2)
```
